```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
 iris = pd.read_csv('/content/Iris (1).csv')
```

```python
# Check shape
print("Shape of dataset:", iris.shape)

# List columns
print("Columns:", iris.columns.tolist())

# Preview first 5 rows
print(iris.head())
```

```python
sns.scatterplot(data=iris, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
# Using seaborn for nicer style
sns.histplot(iris['SepalLengthCm'], kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
sns.boxplot(x='Species', y='SepalWidthCm', data=iris)
plt.title('Box Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
sns.pairplot(iris, hue='Species')
plt.suptitle('Pairwise Relationships between Features', y=1.02)
plt.show()
```

```python
summary = iris.groupby('Species').agg({
    'SepalLengthCm': ['mean', 'min', 'max'],
    'SepalWidthCm': ['mean', 'min', 'max'],
    'PetalLengthCm': ['mean', 'min', 'max'],
    'PetalWidthCm': ['mean', 'min', 'max']
})

print(summary)

```
